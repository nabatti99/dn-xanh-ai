```python
from PIL import Image, ImageDraw2
from ultralytics import YOLO
```

```python
classification_model = YOLO("classification.pt")
classification_model
```

```python
predictions = classification_model.predict("image.jpg")
prediction = predictions[0]
prediction
```

```python
detection_model = YOLO("yolo11m.pt")
detection_model
```

```python
predictions = detection_model.predict("image4.jpg", save=True)
prediction = predictions[0]
prediction
```

```python
image = Image.fromarray(prediction.orig_img)
drawed_image = ImageDraw2.Draw(image)

can_recognize = False

image_width, image_height, _ = prediction.orig_img.shape
image_width, image_height

for box in prediction.boxes:
    prediction_name = prediction.names[int(box.cls)]
    print("Prediction name:", prediction_name)
    if prediction_name == "person":
        continue

    x, y, w, h = box.xywh[0]
    drawed_image.rectangle([x-w/2, y-h/2, x+w/2, y+h/2], pen=ImageDraw2.Pen(color="red"))

    if w > image_width * 0.7 or h > image_height * 0.7:
        can_recognize = True

print("Can recognize:", can_recognize)
drawed_image.image
```
